```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(final[["acc_x","acc_y","acc_z","gyr_x","gyr_y","gyr_z","mag_x","mag_y","mag_z"]])
```

```python
df = pd.DataFrame(data = x,columns = ["acc_x","acc_y","acc_z","gyr_x","gyr_y","gyr_z","mag_x","mag_y","mag_z"])
df["label"] = final["label"].values
df.head()
```

```python
import scipy.stats as stats
```

```python
fs = 80
hop_size = 40
import numpy as np
```

```python
def get_frame(df,fs,hop_size):
  n_features = 9
  frames = []
  labels = []
  for i in range(0,len(df)-fs,hop_size):
    x1 = df["acc_x"].values[i:i+fs]
    x2 = df["acc_y"].values[i:i+fs]
    x3 = df["acc_z"].values[i:i+fs]
    x4 = df["gyr_x"].values[i:i+fs]
    x5 = df["gyr_y"].values[i:i+fs]
    x6 = df["gyr_z"].values[i:i+fs]
    x7 = df["mag_x"].values[i:i+fs]
    x8 = df["mag_y"].values[i:i+fs]
    x9 = df["mag_z"].values[i:i+fs]

    label = stats.mode(df["label"][i:i+fs])[0][0]
    frames.append([x1,x2,x3,x4,x5,x6,x7,x8,x9])
    labels.append(label)
  frames = np.asarray(frames).reshape(-1,fs,n_features)
  labels = np.asarray(labels)
  return frames,labels
```

```python
x,y  = get_frame(df,fs,hop_size)
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python
x_train.shape,x_test.shape
```

```python
x_train = x_train.reshape(8572,80,9,1)
x_test = x_test.reshape(3675,80,9,1)
```

**2D CNN**

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPool2D
from tensorflow.keras.layers import Dense,Dropout,Flatten,BatchNormalization
from tensorflow.keras.optimizers import Adam

model = Sequential()
model.add(Conv2D(16, (2,2), activation ="relu",input_shape = x_train[0].shape))
model.add(Dropout(0.1))

model.add(Conv2D(16,(2,2), activation ="relu"))
model.add(Dropout(0.2))

model.add(Flatten())

model.add(Dense(64, activation ="relu"))
model.add(Dropout(0.5))

model.add(Dense(10, activation = "softmax"))
    
```

```python
model.compile(optimizer=Adam(learning_rate = 0.001), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs = 10, validation_data= (x_test, y_test),verbose = 1)
```

```python
def plot_learningCurve(history,epochs):
  range_epoch = range(1,epochs+1)
  plt.plot(range_epoch,history.history["accuracy"])
  plt.plot(range_epoch,history.history["val_accuracy"])
  plt.title("Model Accuracy")
  plt.xlabel("Epochs")
  plt.ylabel("Accuracy")
  plt.legend(["Training","Validation"])
  plt.show()


  plt.plot(range_epoch,history.history["loss"])
  plt.plot(range_epoch,history.history["val_loss"])
  plt.title("Model Loss")
  plt.xlabel("Epochs")
  plt.ylabel("Loss")
  plt.legend(["Training","Validation"])
  plt.show()


```

```python
plot_learningCurve(history,10)
```
